## Exploring Relationships

### Correlations

```python
# Create a correlation matrix using pandas

df_num.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(df_num.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_num.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_num.shape[1]), df_num.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_num.shape[1]), df_num.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
df_num.rename(columns = {'root_mean_square_amplitude_spectrum' : 'rms_amplitude', 'number_stations_reported' : 'num_reported', 'gap_between_stations' : 'station_gap'},
                      inplace = True)
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_num.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_num.shape[1]), df_num.columns, fontsize=13, rotation=45) # x axis labels
plt.yticks(range(df_num.shape[1]), df_num.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_num.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

### Scatterplots

```python
# Create a scatterplot for the "magnitude" and "num_reported" columns in seaborn

sns.lmplot(x = 'magnitude', y = 'num_reported', data = df_num)
```

```python
# Create a scatterplot for the "magnitude" and "station_gap" columns in seaborn

sns.lmplot(x = 'magnitude', y = 'station_gap', data = df_num)
```

```python
# Create a scatterplot for the "magnitude" and "rms_amplitude" columns in seaborn

sns.lmplot(x = 'magnitude', y = 'rms_amplitude', data = df_num)
```

### Pair Plots

```python
# Keep only the variables we want to use in the pair plot

df_sub = df_num[['magnitude', 'depth', 'num_reported', 'rms_amplitude']]
```

```python
df_sub.head()
```

```python
# Create a pair plot 

g = sns.pairplot(df_sub)
```

### Categorical Plots

```python
# Using a histogram to visualize the distribution of the variables. 
sns.histplot(df_sub['magnitude'], bins = 20, kde = True)
```

```python
df_sub.loc[df_sub['magnitude'] < 4, 'Magnitude category'] = 'Low magnitude'
```

```python
df_sub.loc[(df_sub['magnitude'] >= 4) & (df_sub['magnitude'] < 6), 'Magnitude category'] = 'Middle magnitude'
```

```python
df_sub.loc[df_sub['magnitude'] >= 6, 'Magnitude category'] = 'High magnitude'
```

```python
df_sub['Magnitude category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the categories created above

g = sns.catplot(x="num_reported", y="magnitude", hue="Magnitude category", data=df_sub)
```

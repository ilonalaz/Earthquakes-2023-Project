## 1. Cleaning the dataframe.

Since we will not do any Station or Error Analysis we can delete the columns we don't need.

```python
df=df.drop(columns = ['magType','dmin','net', 'updated', 'horizontalError', 'depthError', 'magError','magNst', 'locationSource', 'magSource'])
```

```python
df.head()
```

```python
df.info()
```

```python
#changing data type
df['time'] = pd.to_datetime(df['time'])
df.info()
```

```python
#code to check for columns that have mixed data type
for col in df.columns.tolist():
    weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    
    if len (df[weird]) > 0:
        print (col)
```

```python
df['place'] = df['place'].astype('str')
```

```python
#rename columns names
df.rename(columns={"mag":"magnitude", "nst":"number_stations_reported", "gap":"gap_between_stations", "rms":"root_mean_square_amplitude_spectrum"}, inplace=True)
df.info()
```
